```python
import http.client, urllib.parse, json
from xml.etree import ElementTree
```

## Issue token

Post an empty payload with key in header to get a token. Note: new unified SpeechService API key and issue token uri is per region. Get new unified SpeechService key for [Free](https://azure.microsoft.com/en-us/try/cognitive-services/?api=speech-services) or [Paid](https://go.microsoft.com/fwlink/?LinkId=872236).

```python
apiKey = "Insert your key here"

headers = {"Ocp-Apim-Subscription-Key": apiKey}
conn = http.client.HTTPSConnection("westus.api.cognitive.microsoft.com")
conn.request("POST", "/sts/v1.0/issueToken", "", headers)
response = conn.getresponse()
data = response.read()
conn.close()

accesstoken = data.decode("UTF-8")
print(accesstoken)
```

## Build SSML

Refer to [Speech Synthesis Markup Language (SSML) Version 1.0](http://go.microsoft.com/fwlink/?LinkId=201763)

```python
body = ElementTree.Element('speak', version='1.0')
body.set('{http://www.w3.org/XML/1998/namespace}lang', 'en-us')
voice = ElementTree.SubElement(body, 'voice')
voice.set('{http://www.w3.org/XML/1998/namespace}lang', 'en-US')
voice.set('{http://www.w3.org/XML/1998/namespace}gender', 'Male')
voice.set('name', 'Microsoft Server Speech Text to Speech Voice (en-US, Guy24KRUS)')
voice.text = 'This is a demo to call microsoft text to speech service.'
ssml = ElementTree.tostring(body)
print(ssml.decode("UTF-8"))
```

## Synthesize

```python
headers = {"Content-type": "application/ssml+xml",
           "X-Microsoft-OutputFormat": "riff-24khz-16bit-mono-pcm",
           "Authorization": "Bearer " + accesstoken,
           "X-Search-AppId": "07D3234E49CE426DAA29772419F436CA",
           "X-Search-ClientID": "1ECFAE91408841A480F00935DC390960",
           "User-Agent": "TTSForPython"}
conn = http.client.HTTPSConnection("westus.tts.speech.microsoft.com")
conn.request("POST", "/cognitiveservices/v1", ElementTree.tostring(body), headers)
response = conn.getresponse()
data = response.read()
conn.close()
print(len(data))
```

## Verify
